```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = '/home/mitsukism/qiskit_entropy/vn_net_gradients_bp/'
gradients_dir = os.path.join(path, 'gradients')
os.makedirs(gradients_dir, exist_ok=True)  # 保存先ディレクトリが存在しない場合に作成

for qubits in [8, 10]:
    matrices = []  # ループごとに初期化
    qubits_dir = os.path.join(path, f'qubits{qubits}')
    
    for filepath in os.listdir(qubits_dir):
        if filepath.endswith('layer3.npy'):
            file_full_path = os.path.join(qubits_dir, filepath)
            matrices.append(np.load(file_full_path))
    
    matrices = np.array(matrices)  # ループ後に一度だけ変換
    print(matrices.shape)
    
    output_file = os.path.join(gradients_dir, f'gradients_qubits{qubits}_layer3.npy')
    np.save(output_file, matrices)
```

```python
# フォルダ内の全ての.npyファイルを読み込み、それぞれの行列をリストに格納する
folder_path = '/home/mitsukism/qiskit_entropy/vn_net_gradients_bp/qubits2/'
npy_files = [f for f in os.listdir(folder_path) if f.endswith('.npy')]

# 行列を格納するリスト
matrices = []

# 各.npyファイルから3*3行列を読み込む
for npy_file in npy_files:
    matrix = np.load(os.path.join(folder_path, npy_file))
    matrices.append(matrix)

# 行列のリストをnumpy配列に変換（形状: (ファイル数, 3, 3)）
matrices = np.array(matrices)

# 各成分の平均と標準偏差を計算
mean_matrix = np.mean(matrices, axis=0)
std_matrix = np.std(matrices, axis=0)

# 平均と標準偏差を三次元プロットする準備
x, y = np.meshgrid(range(3), range(3))

# 3Dプロットの作成
fig = plt.figure(figsize=(18, 6))

# 平均のプロット
ax1 = fig.add_subplot(121, projection='3d')
ax1.bar3d(x.flatten(), y.flatten(), np.zeros_like(mean_matrix.flatten()), 1, 1, mean_matrix.flatten())
ax1.set_title('Mean of Matrix Elements qubits2')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Mean')

# 標準偏差のプロット
ax2 = fig.add_subplot(122, projection='3d')
ax2.bar3d(x.flatten(), y.flatten(), np.zeros_like(std_matrix.flatten()), 1, 1, std_matrix.flatten())
ax2.set_title('Standard Deviation of Matrix Elements qubits2')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_zlabel('Standard Deviation')

# グラフを表示
plt.show()
```

```python
qubits_list = [2, 6, 8, 10]
layers_list = [3, 5, 7, 9, 11]
path = '/home/mitsukism/qiskit_entropy/vn_net_gradients_bp/'

plt.figure(figsize=(12, 6))

for layers in layers_list:
    grad_vars = []
    for qubits in qubits_list:
        matrices = np.load(path + f'gradients/gradients_qubits{qubits}_layer{layers}.npy')
        var_matrix = np.var(matrices, axis=0)
        indices = [int(layers/2), 1]
        grad_vars.append(var_matrix[tuple(indices)])
    plt.plot(qubits_list, grad_vars, marker='o', label=f'Layer {layers}')

plt.yscale('log')
plt.title('Variance of Gradient Elements')
plt.xlabel('Number of Qubits')
plt.ylabel('Variance $\\frac{\\partial E}{\\partial \\theta}$')
plt.legend()

plt.figure(figsize=(12, 6))

for qubits in qubits_list:
    grad_vars = []
    for layers in layers_list:
        matrices = np.load(path + f'gradients/gradients_qubits{qubits}_layer{layers}.npy')
        var_matrix = np.var(matrices, axis=0)
        indices = [int(layers/2), 1]
        grad_vars.append(var_matrix[tuple(indices)])
    plt.plot(layers_list, grad_vars, marker='o', label=f'qubit {qubits}')

plt.yscale('log')
plt.title('Variance of Gradient Elements')
plt.xlabel('Number of Layers')
plt.ylabel('Variance $\\frac{\\partial E}{\\partial \\theta}$')
plt.legend()
```
